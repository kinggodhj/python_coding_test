```python
import pandas as pd
import numpy as np
```

```python
path='./analysis_2017.csv'
data=pd.read_csv(path)
```

```python
#결측값 확인
data.isnull().sum()
```

```python
#변수들 타입 확인
data.dtypes
```

```python
#normalization
for d in data.columns:
    if d=='연봉(2018)':
        continue
    if data[d].dtypes=='int64' or data[d].dtypes=='float64':
        mean=data[d].mean()
        std=data[d].std()
        data[d]=data[d].apply(lambda x:(x-mean)/std)
```

```python
data=data.rename(columns={'연봉(2018)':'y'})
```

```python
onehot=pd.get_dummies(data['팀명'])
```

```python
data.join(onehot)
```

```python
data=data.drop(['선수명'], axis=1)
data=data.drop(['팀명'], axis=1)
```

```python
data
```

```python
#train data, test data
y=data['y']
x=data[data.columns.difference(['y'])]
```

```python
x
```

```python
y
```

## Data split

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data, train_label, test_label=train_test_split(x,y, test_size=0.3, random_state=19)
```

```python
train_data
```

```python
train_label
```

## Model train

```python
m=linear_model.LinearRegression()
```

```python
model=m.fit(train_data, train_label)
```

```python
pred=model.predict(test_data)
```

```python
pd.DataFrame({'pred':pred, 'target':test_label})
```

## R2 score

```python
print(model.score(train_data, train_label))
print(model.score(test_data, test_label))
```
